```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('fivethirtyeight')
future_bed_site=769.4
bed_site=769.4
price=2200
butie=186
def cost(price,butie,bed_site,future_bed_site):
    return (price+butie)/np.log10(future_bed_site/bed_site*10)
cost(price,butie,bed_site,future_bed_site)
```

```python
x=[]
y=[]
for i in range(100):
    future_bed_site=bed_site*(1+i)/10
    x.append(i/10)
    y.append(cost(price,butie,bed_site,future_bed_site))
plt.plot(x,y,'-',label='cost')
plt.legend(loc='upper right')
```

```python
def single_butie(butie,bed_site,future_bed_site):
    return butie*bed_site/future_bed_site
x=[]
y=[]
for i in range(20):
    future_bed_site=bed_site*(i+1)/10
    x.append(i/10)
    y.append(cost(price,butie,bed_site,future_bed_site)-single_butie(butie,bed_site,future_bed_site))
plt.plot(x,y,'o-',label='cost')
plt.hlines(2076, 0, 2,color="red",linewidth=4,label='y=2076')#竖线
plt.legend(loc='upper right')
```
